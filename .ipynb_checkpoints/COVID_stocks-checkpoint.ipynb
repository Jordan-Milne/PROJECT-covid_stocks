## Input Symbol Below

```python
symbol = 'AAPL'
```

## Now all you have to do is run the rest of the code!

```python
# !pip install alpha_vantage
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from matplotlib.ticker import FuncFormatter
%config InlineBackend.figure_format='retina'

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
NASDAQ = yf.Ticker("^IXIC").history(period="1y")
NASDAQ['date'] = NASDAQ.index
NASDAQ = NASDAQ.reset_index(drop=True)
```

```python
data = yf.Ticker(symbol).history(period="1y")
data['date'] = data.index
data = data.reset_index(drop=True)
```

```python
def stock_info(data):
    df1 = data[len(data[data['date'] < '2020-02-11'])+1:]
    percent = round((df1['Close'].iloc[-1]/df1['Close'].iloc[0]-1)*100,2)
    print(f'Percent down since COVID-19 market crash (February 12 2020) = {percent}%')
```

```python
def prior(stock_data):
    mask1 = stock_data['date'] >= '2019-08-12'
    mask2 = stock_data['date'] <= '2020-02-12'
    dff = stock_data[mask1 & mask2]
    
    plt.figure(figsize=(8,5))
    plt.plot(dff['date'],dff['Close'])
    plt.tight_layout()
    plt.title("6 Months Prior to COVID-19 Market Effects")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.grid();
```

```python
def comparison(stock_data,stock_symbol):
    mask1 = data['date'] >= '2019-08-12'
    mask2 = data['date'] <= pd.to_datetime('today')
    dff = data[mask1 & mask2]
    dff2 = NASDAQ[mask1 & mask2]
    
    fig, ax = plt.subplots(1,figsize=(10,5))
    ax.plot(dff['date'],dff['Close']/dff['Close'].iloc[0]-1)
    ax.plot(dff2['date'],dff2['Close']/dff2['Close'].iloc[0]-1)
    ax.set(xlabel='Date', title=stock_symbol+' Compared to NASDAQ (^IXIC)')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
    ax.grid();
```

```python
stock_info(data)
prior(data)
comparison(data, symbol)
```
